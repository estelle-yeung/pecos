```python
# lufkin

ns_l = lufkin[lufkin['Direction'].isin(['North','South'])].Count.sum()
we_l = lufkin[lufkin['Direction'].isin(['West','East'])].Count.sum()
nesw_l = lufkin[lufkin['Direction'].isin(['Northeast','Southwest'])].Count.sum()
nwse_l = lufkin[lufkin['Direction'].isin(['Northwest','Southeast'])].Count.sum()

lufkin_combined = pd.DataFrame({'direction':['n+s', 'w+e', 'ne+sw', 'nw+se'], 'count':[ns_l, we_l, nesw_l, nwse_l]})
```
