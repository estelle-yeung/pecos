```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
raw = pd.read_csv('wind_pecos_20200401_20240318.csv')
```

```python
raw.head()
```

```python
raw = raw.rename(columns={'wind_ws_ms':'wind_we_ms'})
```

`wind_we_ms` describes the wind in eastward direction in meter per second. The positive number means east wind (to the west) and the negative number means west wind (to the east).

`wind_ns_ms` describes the wind in northward direction in meter per second. The positive number means north wind (to the south) and the negative number means south wind (to the north).

```python
type(raw.time[1])
```

```python
raw['date_format'] = pd.to_datetime(raw['time'], format='%Y-%m-%dT%H:%M:%SZ').dt.date
raw['time_format'] = pd.to_datetime(raw['time'], format='%Y-%m-%dT%H:%M:%SZ').dt.time
```

```python
raw['month_format'] = pd.to_datetime(raw['time'], format='%Y-%m-%dT%H:%M:%SZ').dt.month
raw['year_format'] = pd.to_datetime(raw['time'], format='%Y-%m-%dT%H:%M:%SZ').dt.year
```

```python
raw.head()
```

```python
raw.groupby('time_format')['wind_we_ms'].mean()
```

```python
raw.groupby('time_format')['wind_ns_ms'].mean()
```

```python
raw.groupby('month_format')['wind_we_ms'].mean()
```

```python
raw.groupby('month_format')['wind_ns_ms'].mean()
```

```python
raw.describe()
```

```python
raw.groupby(['month_format','time_format'])['wind_we_ms'].mean()
```

```python
raw.groupby(['year_format','month_format'])['wind_we_ms'].mean()
```
